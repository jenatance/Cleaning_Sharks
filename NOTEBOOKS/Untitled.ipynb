```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')

import pylab as plt
import seaborn as sns

%matplotlib inline
```

```python
sharks = pd.read_csv('../DATA/attacks.csv', encoding = "ISO-8859-1")
```

```python
sharks = sharks.dropna(how = 'all') # eliminamos las filas que tengan todo nulo
```

```python
sharks.shape, sharks_ori_shape # tamaño del df ahora vs al inicio
```

```python
sharks = sharks[sharks.isna().sum(axis=1) < 10]
```

```python
# Las columnas Unnamed: 22 y Unnamed: 23 tendríamos que rellenarlas con datos, en este caso, al ser de tipo Object, con un String 'desconocido'
```

```python
sharks[['Unnamed: 22', 'Unnamed: 23']] = sharks[['Unnamed: 22', 'Unnamed: 23']].fillna('Unknown')
```

```python
nanCols = sharks.isna().sum()
nanCols[nanCols > 0]
```

```python
columns_names = sharks.columns.values
columns_names
```

```python
sharks.rename(columns={'Case Number': 'caseNumber', 'Date': 'date', 'Year': 'year', 'Type': 'type', 'Country': 'country', 'Area': 'area',
       'Location': 'location', 'Activity': 'activity', 'Name': 'name', 'Sex ': 'sex', 'Age': 'age', 'Injury': 'injury',
       'Fatal (Y/N)': 'fatal(y/n)', 'Time': 'time', 'Species ': 'species', 'Investigator or Source': 'investigatorOrSource', 'pdf': 'pdf',
       'href formula': 'hrefFormula', 'href': 'href', 'Case Number.1': 'caseNumber1', 'Case Number.2': 'caseNumber2',
       'original order': 'originalOrder', 'Unnamed: 22': 'unnamed22', 'Unnamed: 23': 'unnamed23'}, inplace=True)
```

```python
print(sharks['year'].unique())
```

```python
print(sharks['sex'].unique())
```

```python
sharks.shape, sharks_ori_shape # tamaño del df ahora vs al inicio
```

```python
indexNames = sharks[ sharks['year'] < 2001 ].index # índices de los ataques antes de 2002
sharks.drop(indexNames, inplace=True) # quitamos de nuestro dataframe los índices que hemos hallado previamente
indexNames = sharks[ sharks['sex'] == 'F' ].index # índices de los ataques a mujeres
sharks.drop(indexNames, inplace=True) # quitamos de nuestro dataframe los índices que hemos hallado previamente
indexNames = sharks[ sharks['sex'] == 'lli' ].index # índices de los ataques a otros
sharks.drop(indexNames, inplace=True) # quitamos de nuestro dataframe los índices que hemos hallado previamente
```

```python
sharks = sharks[sharks['year'].notna()]
sharks = sharks[sharks['sex'].notna()]
```

```python
sharks.sex = sharks.sex.str.replace('M ', 'M') # para poner todos los datos de género con el mismo formato
```

```python
print(sharks['year'].unique())
```

```python
print(sharks['sex'].unique())
```

```python
sharks.shape, sharks_ori_shape # tamaño del df solo con los datos de los ataques de los últimos 50 años (1972-2022) vs al inicio
```

```python
nanCols = sharks.isna().sum()
nanCols[nanCols > 0]
```

```python
print(sharks['type'].unique())
```

```python
sharks.type = sharks.type.str.replace('Boating', 'Boat')
sharks.type = sharks.type.str.replace('Boatomg', 'Boat')
sharks.type = sharks.type.str.replace('Boat', 'Boating')
```

```python
print(sharks['type'].unique())
```

```python
display(sharks[sharks['type'].isnull()]) # para ver qué ocurre con los valores nulos de type
```

```python
sharks.loc[85,'type'] = ['Unprovoked']
sharks.loc[382,'type'] = ['Unprovoked']
```

```python
print(sharks['country'].unique())
```

```python
sharks['country'] = sharks['country'].str.lower() 
```

```python
display(sharks[sharks['country'].isnull()]) 
```

```python
sharks.loc[62,'country'] = ['Unknown']
sharks.loc[525,'country'] = ['Unknown']
```

```python
sharks[['name']] = sharks[['name']].fillna('Unknown')
```

```python
print(sharks['age'].unique())
```

```python
sharks[['age']] = sharks[['age']].fillna('-1')
```

```python
sharks.age = sharks.age.str.replace('40s', '40')
sharks.age = sharks.age.str.replace('Teen', '15')
sharks.age = sharks.age.str.replace('60s', '60')
sharks.age = sharks.age.str.replace('18 months', '0')
sharks.age = sharks.age.str.replace('50s', '50')
sharks.age = sharks.age.str.replace('teen', '15')
sharks.age = sharks.age.str.replace('28 & 26', '27')
sharks.age = sharks.age.str.replace('30s', '30')
sharks.age = sharks.age.str.replace('46 & 34', '40')
sharks.age = sharks.age.str.replace('15s', '15')
sharks.age = sharks.age.str.replace('\xa0 ', '-1')
sharks.age = sharks.age.str.replace('30 or 36', '33')
sharks.age = sharks.age.str.replace('6½', '6')
sharks.age = sharks.age.str.replace('21 & ?', '21')
sharks.age = sharks.age.str.replace('33 or 37', '35')
sharks.age = sharks.age.str.replace('mid-30', '35')
sharks.age = sharks.age.str.replace('12 or 13', '12')
```

```python
sharks.age = sharks.age.str.replace(' ', '-1')
```
