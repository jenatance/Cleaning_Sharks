```python
import pandas as pd
pd.set_option('display.max_columns', None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')

import pylab as plt
import seaborn as sns

%matplotlib inline
```

```python
sharks = pd.read_csv('../DATA/attacks.csv', encoding = "ISO-8859-1")
```

```python
sharks.head()
```

```python
sharks_ori_shape = sharks.shape
sharks_ori_shape
```

```python
sharks.tail(20)  # podemos observar que al final de nuestro dataframe, hay muchas filas que no tienen ningún dato
```

```python
sharks = sharks.dropna(how = 'all') # eliminamos las filas que tengan todo nulo
```

```python
sharks.shape, sharks_ori_shape # tamaño del df ahora vs al inicio
```

```python
sharks # vemos que todavía tenemos muchas filas con muchos nulos
```

```python
sharks = sharks[sharks.isna().sum(axis=1) < 10]
```

```python
sharks
```

```python
sharks.shape, sharks_ori_shape # tamaño del df ahora vs al inicio
```

```python
sharks.info() # vemos las columnas con su tipo de variable y los no nulos que tienen cada una
```

```python
# Las columnas Unnamed: 22 y Unnamed: 23 tendríamos que rellenarlas con datos, en este caso, al ser de tipo Object, con un String 'desconocido'
```

```python
sharks[['Unnamed: 22', 'Unnamed: 23']] = sharks[['Unnamed: 22', 'Unnamed: 23']].fillna('desconocido')
```

```python
nanCols = sharks.isna().sum()
nanCols[nanCols > 0]
```

```python
nanCols = sharks.isna().sum()
nanCols[nanCols > 1000]
```

```python
# Mirando los valores no nulos, deberíamos mirar la columna 'age', la columna 'time' y la columna 'species'
```
